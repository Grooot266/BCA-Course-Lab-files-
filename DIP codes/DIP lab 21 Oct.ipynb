```python
#hough transform circle detection
import os
import cv2
import numpy as np
c=0
img = cv2.imread('D:/IMG processing lab/44ae9f3db5c855bf0703058fba4f44fc.jpg',0)
cv2.imshow("inp:-",img)
img1 = cv2.blur(img, (3, 3))
detected_circles = cv2.HoughCircles(img1,
cv2.HOUGH_GRADIENT, 1, 20, param1 = 50,
param2 = 30, minRadius = 1, maxRadius = 40)
if detected_circles is not None:
    detected_circles = np.uint16(np.around(detected_circles))

    for pt in detected_circles[0, :]:
        a, b, r = pt[0], pt[1], pt[2]
        cv2.circle(img1, (a, b), r, (0, 255, 0), 2)
        cv2.circle(img1, (a, b), 1, (0, 0, 255), 3)
        c=c+1
    cv2.imshow("Circle detection:-", img1)
    cv2.waitKey(0)
    print(c)
    
```

```python
#roi detction bottle problem
import cv2
import numpy as np
img = cv2.imread('D:/IMG processing lab/MicrosoftTeams-image (7).png',0)
prep = img[30:900,258:510]
cv2.imshow('ROI img:-',prep)

edges = cv2.Canny(prep,60,200)

cv2.imshow('Edge:-',edges)
cv2.waitKey(0)

# img.shape
r=img.shape[0]
print("rows:-",r)
c=img.shape[1]
print("columns:-",c)


er=edges.shape[0]
print("Edge row:-",er)
ec=edges.shape[1]
print("Edge column:-",ec)
count=0
ar=[]

for i in range(edges.shape[0]):
    for j in range(edges.shape[1]):
        if (edges[i][j]==255):
            ar.append(i)
            count=count+1
            print("[",i,"]","[",j,"]=",edges[i][j])
            
print(count)


        

```
