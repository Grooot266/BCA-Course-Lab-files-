Q.1	Serach and replace all dates in the text given with Saturday
Text= Ram played Guitar on 12-09-1992 and he was admitted to school on 15-12-1999" .


```python
#search :-
import re
string="Ram played Guitar on 12-09-1992 and he was admitted to school on 15-12-1999"
pattern="[0-9]{2}[-|\/]{1}[0-9]{2}[-|\/]{1}[0-9]{4}"
result=re.findall(pattern,string)
print(result)
```

```python
#replace:-
pattern="[0-9]{2}[-|\/]{1}[0-9]{2}[-|\/]{1}[0-9]{4}"
replacement="Saturday"
result=re.sub(pattern,replacement,string)
print(result)
```

Q.2	Given a text, perform lemmatization.
text = "Walter was feeling anxious. He was diagnosed today. He probably is the best person I know."


```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
text ="Walter was feeling anxious. He was diagnosed today. He probably is the best person I know."
lemm=WordNetLemmatizer()
print("Word list of the text:-")
word_list = nltk.word_tokenize(text)
print(word_list)
print("\n")
print("Lemmatization:-")
for w in word_list:
    output=lemm.lemmatize(w)
    print(output)
```

Q.3	Find POS tags for the text given below also display chunks by creating own rule.
Text=”Donald John Trump is an American media personality and businessman who served as the 45th president of the United States from 2017 to 2021.”


```python
import nltk
from nltk import pos_tag
from nltk import RegexpParser
from nltk.tokenize import word_tokenize
text= "Donald John Trump is an American media personality and businessman who served as the 45th president of the United States from 2017 to 2021."
tokens=word_tokenize(text)
mo=pos_tag(tokens)
print("Pos-tags:-")
print(mo)
pattern = """mychunk:{<NN.?>*<VB.?>*<JJ.?>*<V.*>*<CC>?}"""
chunks = RegexpParser(pattern)
print("After Regex:-",chunks)
op=chunks.parse(mo)
print("After chunking:-",op)
```

Q.4	Find tf-idf value for the text given below:
sentences = ['This is the first document','This document is the second document']


```python
from sklearn.feature_extraction.text import TfidfVectorizer
sentences = ['This is the first document','This document is the second document']
vectorizer = TfidfVectorizer(norm = None)
X= vectorizer.fit_transform(sentences).toarray()
print(vectorizer.vocabulary_)
print("\n")
print(vectorizer.get_feature_names())
print("\n")
print(X)
```
